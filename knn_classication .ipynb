```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.head()
```

```python
x=df.drop(["target"],axis=1)
y=df["target"]
```

```python
x
```

```python
y
```

### MinMaxScaler

```python
normal_scaler=MinMaxScaler(feature_range=(0,1))
x_scaled=normal_scaler.fit_transform(x)
x_scaled
x_normal_df=pd.DataFrame(x_scaled,columns=x.columns)
x_normal_df
```

### Standardscaler

```python
std_scaler=StandardScaler()
x_scaled=std_scaler.fit_transform(x)
x_scaled
x_std_df=pd.DataFrame(x_scaled,columns=x.columns)
x_std_df
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_normal_df,y,test_size=0.2,random_state=1,stratify=y)
x_train
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
np.arange(2,20)
```

```python
param_grid={"n_neighbors":np.arange(2,10),
           "p":[1,2]}
param_grid
```

```python
knn_clf=KNeighborsClassifier()
knn_clf
```

```python
gscv=GridSearchCV(knn_clf,param_grid,cv=10)
gscv.fit(x_train,y_train)
```

```python
gscv.best_estimator_
```

```python
gscv.best_params_
```

```python
rscv=RandomizedSearchCV(knn_clf,param_grid,cv=5)
rscv.fit(x_train,y_train)
```

```python
rscv.best_estimator_
```

```python
rscv.best_params_
```

```python
knn_clf=KNeighborsClassifier(n_neighbors=6,p=1)
knn_clf.fit(x_train,y_train)
```

```python
y_pred=knn_clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
y_pred[:5]
```

```python
y_test[:5]
```

```python
confusion_matrix(y_test,y_pred)
```

```python
clf=classification_report(y_test,y_pred)
print(clf)
```

```python
accuracy_score(y_test,y_pred)
```
